```python
#Load dataset
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/P_Data_Extract_From_World_Development_Indicators.xlsx', sheet_name='Data')
```

```python
#We have columns that we do not need for this analysis
df = df.drop(columns = ['Series Code', 'Country Name', 'Country Code', '2023 [YR2023]'])
df
```

```python
#We do not need the last 5 rows
indices_to_drop = [14, 15, 16, 17, 18]
df = df.drop(indices_to_drop)
df
```

```python
df.info()
```

As we can see Dtype for all columns is "object". To perform other calculations and visualize the data we need to convert data type for each columns to "float"

```python
# Reset index to get the row names back as columns
df = df.reset_index()

# The first row after transposition will be the new headers
df.columns = df.iloc[0]

# Drop the first row which is now redundant as headers
df = df.drop(df.index[0])

# Convert the data to float, skipping the first column which contains the new headers
df.iloc[:, 1:] = df.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

# Reset the index to make it a proper DataFrame again
df.reset_index(inplace=True, drop=True)

df.head()
```

```python
df.info()
```

```python
# Convert specific columns to float
columns_to_convert = [
    'GDP (current US$)', 'Exports of goods and services (% of GDP)',
    'Exports of goods and services (annual % growth)', 'Exports of goods and services (current US$)',
    'GDP growth (annual %)', 'GDP per capita (current US$)',
    'GDP per capita growth (annual %)', 'Imports of goods and services (% of GDP)',
    'Imports of goods and services (annual % growth)', 'Imports of goods and services (current US$)',
    'GNI (current US$)', 'GNI growth (annual %)',
    'GNI per capita growth (annual %)', 'GNI per capita, Atlas method (current US$)'
]

for col in columns_to_convert:
    df[col] = df[col].astype(float)

df.info()
```

```python
# Display the DataFrame before converting to billions
print("DataFrame before converting to billions:")
print(df.head())
```

```python
# Convert values to billions and round them
columns_to_convert_to_billions = [
    'GDP (current US$)', 'Exports of goods and services (current US$)',
    'Imports of goods and services (current US$)', 'GNI (current US$)'
]

for col in columns_to_convert_to_billions:
    df[col] = (df[col] / 1e9).round(2)
    new_col_name = f"{col} (in billions)"
    df.rename(columns={col: new_col_name}, inplace=True)
    print(f"Converted to billions: {new_col_name}")
    print(df[new_col_name].head())
```

```python
df.head()
```

```python
#round values to 2 digits after decimal point
columns_to_round = [
    'Exports of goods and services (% of GDP)', 'Exports of goods and services (annual % growth)',
    'GDP growth (annual %)', 'GDP per capita (current US$)',
    'GDP per capita growth (annual %)', 'Imports of goods and services (% of GDP)',
    'Imports of goods and services (annual % growth)', 'GNI growth (annual %)',
    'GNI per capita growth (annual %)'
]

for col in columns_to_round:
    df[col] = df[col].round(2)

df.head()
```

We can see that column "GNI per capita growth (annual %)" has one Null-value. This column represents annual growth of Gross National Income per capita. Based on information from column "GNI (current US$)" we can calculate annual growth that missing.

```python
#First, we need to find in what year we have missing value
missing_gni_per_capita_growth = df[df['GNI per capita growth (annual %)'].isna()]
missing_gni_per_capita_growth
```

```python
new_gni_per_capita_growth_2022 = df.loc[22, 'GNI per capita, Atlas method (current US$)'] / df.loc[21, 'GNI per capita, Atlas method (current US$)']
new_gni_per_capita_growth_2022 = new_gni_per_capita_growth_2022.round(2)
new_gni_per_capita_growth_2022
```

```python
missing_value = df.loc[22, 'GNI per capita growth (annual %)']
df.replace (missing_value, new_gni_per_capita_growth_2022, inplace = True)
df.tail()
```

```python
df.info()
```

For analysis we need a foreign trade quota, also known as the trade-to-GDP ratio. It is an economic metric that measures the openness of an economy to international trade.
We can calculate it using folowing formula:
- Foreign Trade Quota= (Total Trade Turnover (Exports + Imports) / Gross Domestic Product (GDP)) Ã— 100

```python
df['Foreign Trade Quota (%)'] = (
    (df['Exports of goods and services (current US$) (in billions)'] +
     df['Imports of goods and services (current US$) (in billions)']) /
    df['GDP (current US$) (in billions)']) * 100

df['Foreign Trade Quota (%)'] = df['Foreign Trade Quota (%)'].round(2)
df.head()
```

```python
# Line plot for GDP (current US$) over time with secondary y-axis for GDP growth (annual %)
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot GDP (current US$) on the primary y-axis
sns.lineplot(ax=ax1, x=df['Series Name'], y=df['GDP (current US$) (in billions)'], color='blue')
ax1.set_title('GDP (current US$) and GDP Growth (annual %) Over Time')
ax1.set_xlabel('Year')
ax1.set_ylabel('GDP (current US$) (in billions)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
plt.xticks(rotation=45)

# Create a secondary y-axis to plot GDP growth (annual %)
ax2 = ax1.twinx()
sns.lineplot(ax=ax2, x=df['Series Name'], y=df['GDP growth (annual %)'], color='orange')
ax2.set_ylabel('GDP Growth (annual %)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

plt.show()
```

We observe a steady GDP growth from 2000 to 2007. In 2008, there is a decline that continues into 2009, linked to the financial crisis of 2008. From 2009 to 2014, GDP starts to rise again. Between 2014 and 2017, there is another decline. In 2018, GDP increases, but it falls again in 2019. This decline can be attributed to several factors such as Brexit uncertainty, a global economic slowdown, and domestic political uncertainty. This marked the most significant decline in the UK's GDP in history.

Based on the line chart, it's clear that the UK's GDP experienced various fluctuations over the years. The sharp drop in 2019 is particularly notable, illustrating the severe impact of the mentioned factors. Additionally, the recovery seen in 2021 indicates resilience and the potential for economic rebound following significant disruptions such as Covid-19.

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
#Dashboard for Export and Import
# Initialize the Dash app
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Import and Export Dashboard"),
    html.Div([
        html.Label("Select Year:"),
        dcc.Dropdown(
            id='year-dropdown',
            options=[{'label': year, 'value': year} for year in df['Series Name']],
            multi=True,
            value=df['Series Name'].unique()
        )
    ]),
    html.Div([
        html.Div([
            dcc.Graph(id='export-chart')
        ], style={'width': '48%', 'display': 'inline-block'}),
        html.Div([
            dcc.Graph(id='import-chart')
        ], style={'width': '48%', 'display': 'inline-block'}),
    ]),
    html.Div([
        dcc.Graph(id='gdp-chart')
    ], style={'width': '100%', 'display': 'inline-block'})
])

@app.callback(
    [Output('export-chart', 'figure'),
     Output('import-chart', 'figure'),
     Output('gdp-chart', 'figure')],
    [Input('year-dropdown', 'value')]
)
def update_charts(selected_years):
    filtered_df = df[df['Series Name'].isin(selected_years)]

    # Export chart with secondary y-axis
    fig_export = go.Figure()
    fig_export.add_trace(
        go.Scatter(x=filtered_df['Series Name'], y=filtered_df['Exports of goods and services (current US$) (in billions)'],
                   name='Exports of goods and services (current US$) (in billions)', yaxis='y1', mode='lines')
    )
    fig_export.add_trace(
        go.Scatter(x=filtered_df['Series Name'], y=filtered_df['Exports of goods and services (annual % growth)'],
                   name='Exports of goods and services (annual % growth)', yaxis='y2', mode='lines')
    )
    fig_export.update_layout(
        title='Export of Goods and Services',
        yaxis=dict(title='Exports of goods and services (current US$) (in billions)'),
        yaxis2=dict(title='Exports of goods and services (annual % growth)', overlaying='y', side='right'),
        legend=dict(x=0, y=1.2, orientation='h')
    )

    # Import chart with secondary y-axis
    fig_import = go.Figure()
    fig_import.add_trace(
        go.Scatter(x=filtered_df['Series Name'], y=filtered_df['Imports of goods and services (current US$) (in billions)'],
                   name='Imports of goods and services (current US$) (in billions)', yaxis='y1', mode='lines')
    )
    fig_import.add_trace(
        go.Scatter(x=filtered_df['Series Name'], y=filtered_df['Imports of goods and services (annual % growth)'],
                   name='Imports of goods and services (annual % growth)', yaxis='y2', mode='lines')
    )
    fig_import.update_layout(
        title='Import of Goods and Services',
        yaxis=dict(title='Imports of goods and services (current US$) (in billions)'),
        yaxis2=dict(title='Imports of goods and services (annual % growth)', overlaying='y', side='right'),
        legend=dict(x=0, y=1.2, orientation='h')
    )

    # GDP chart with secondary y-axis for GDP growth (annual %)
    fig_gdp = go.Figure()
    fig_gdp.add_trace(
        go.Scatter(x=filtered_df['Series Name'], y=filtered_df['GDP (current US$) (in billions)'],
                   name='GDP (current US$) (in billions)', yaxis='y1', mode='lines')
    )
    fig_gdp.add_trace(
        go.Scatter(x=filtered_df['Series Name'], y=filtered_df['GDP growth (annual %)'],
                   name='GDP growth (annual %)', yaxis='y2', mode='lines')
    )
    fig_gdp.update_layout(
        title='GDP (current US$) and GDP Growth (annual %)',
        yaxis=dict(title='GDP (current US$) (in billions)'),
        yaxis2=dict(title='GDP growth (annual %)', overlaying='y', side='right'),
        legend=dict(x=0, y=1.2, orientation='h')
    )

    return fig_export, fig_import, fig_gdp

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```

- 2000-2007: Both exports and imports of goods and services in the UK experienced steady growth, indicating a period of economic stability and expansion.

- 2008-2009: The financial crisis led to significant declines in both exports and imports, with negative growth rates reflecting economic contraction.

- 2009-2014: A period of recovery for both exports and imports, with positive growth rates despite some volatility.

- 2014-2017: Fluctuations and relative stabilization in trade activities, with growth rates showing variability.

- 2018-2019: Brexit-related uncertainty caused significant volatility in both export and import activities.

- 2020: The COVID-19 pandemic led to sharp declines in trade activities, with negative growth rates for both exports and imports.

- 2021-2022: Strong recovery in both exports and imports, with sharp increases in growth rates indicating a robust post-pandemic economic rebound.
